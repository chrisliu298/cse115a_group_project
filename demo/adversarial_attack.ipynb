```
import torch
from pprint import pprint
from transformers import GPT2LMHeadModel, GPT2Tokenizer

from generator import GPT2Generator

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")   

gpt2_generator = GPT2Generator(device)
model = GPT2LMHeadModel.from_pretrained("/content/drive/My Drive/models/gpt2_yelp")
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
_ = model.to(device)
```

```
prompt = "[positive] <|sep|> The restaurant is bad"
pprint(gpt2_generator.generate(
    tokenizer, 
    model, 
    prompt, 
    length=50,
    seed=44)[0]
)
```

```
prompt = "[negative] <|sep|> The restaurant is good"
pprint(gpt2_generator.generate(
    tokenizer, 
    model, 
    prompt, 
    length=100,
    seed=44)[0]
)
```

```
from trl.gpt2 import GPT2HeadWithValueModel
from trl.gpt2 import respond_to_batch

from generator import GPT2PPOGenerator

gpt2_ppo_generator = GPT2PPOGenerator(device)
model = GPT2HeadWithValueModel.from_pretrained("/content/drive/My Drive/models/gpt2_ppo_yelp")
tokenizer = GPT2Tokenizer.from_pretrained("/content/drive/My Drive/models/gpt2_ppo_yelp")
_ = model.to(device)
```

```
prompt = "[positive] The restaurant is bad"
pprint(gpt2_ppo_generator.generate(
    tokenizer, 
    model, 
    prompt,
    length=50)
)
```

```
prompt = "[negative] The restaurant is good"
pprint(gpt2_ppo_generator.generate(
    tokenizer, 
    model, 
    prompt,
    length=50)
)
```
