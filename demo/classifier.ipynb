```
import os
os.chdir("/content/tapt/src")
```

```
# Model and tokenizer
from transformers import AutoModelForSequenceClassification
from transformers import AutoTokenizer
# Classifier
from classifier import Classifier

# Load model
model = AutoModelForSequenceClassification.from_pretrained(
    "/content/drive/My Drive/models/distilroberta_yelp"
)
# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained(
    "/content/drive/My Drive/models/distilroberta_yelp", 
    use_fast=True
)
# Create classifier
clf = Classifier(model=model, tokenizer=tokenizer)
```

```
print(clf.classify("The restaurant is really bad"))
```

```
print(clf.classify("The restaurant is really good"))
```
