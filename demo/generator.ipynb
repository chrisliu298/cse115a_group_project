```
import torch
from pprint import pprint
from transformers import GPT2LMHeadModel, GPT2Tokenizer
# Generator
from generator import GPT2Generator

# Use GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")   

# Create generator
gpt2_generator = GPT2Generator(device)

# Load model
model = GPT2LMHeadModel.from_pretrained(
    "/content/drive/My Drive/models/gpt2_imdb"
)
# Load tokenizer
tokenizer = GPT2Tokenizer.from_pretrained("/content/drive/My Drive/models/gpt2_imdb")
# Move model to GPU
_ = model.to(device)
```

```
prompt = "[positive] <|sep|> The movie is actually"
pprint(gpt2_generator.generate(
    tokenizer, 
    model, 
    prompt, 
    length=50,
    seed=44)[0]
)
```

```
prompt = "[negative] <|sep|> The movie is really"
pprint(gpt2_generator.generate(
    tokenizer, 
    model, 
    prompt, 
    length=100,
    seed=44)[0]
)
```
